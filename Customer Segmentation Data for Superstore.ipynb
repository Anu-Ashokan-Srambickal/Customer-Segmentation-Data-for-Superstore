### üìï ABOUT DATASET

The superstore_data.csv dataset contains information on 2240 customers across 22 columns. 
* A quick overview of the data reveals the following:

### üî∫ Customer Demographics:
* The dataset includes columns for Id, Year_Birth, Education, and Marital_Status.

### üî∫ Financial Information: 
* The Income column represents the customer's income and has 24 missing values.

### üî∫ Family Details: 
* The Kidhome and Teenhome columns show the number of children and teenagers in the household, respectively.

### üî∫ Customer History: 
* The Dt_Customer column records the date of customer enrollment.

### üî∫ Purchasing Behavior:
* There are several columns related to purchasing, including Recency (days since the last purchase) and a series of columns prefixed with Mnt (e.g., MntWines) that represent spending on various products.

### üî∫ Interaction with the Store: 
* The data also contains columns like NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases, and NumWebVisitsMonth, which show how customers interact with the store.

### üî∫  Engagement and Complaints:
* The Response and Complain columns provide information on customer campaign responses and complaints.

* The column data types are primarily integer (int64) for most of the numerical data, with the exception of Income, which is a float.
* The Education, Marital_Status, and Dt_Customer columns are stored as objects.

### üü• ANALYSING AVERAGE SPENDING BASED ON EDUCATION AND NUMBER OF DEALS

```python
# Group by 'NumDealsPurchases' and 'Education' and calculate the mean of 'MntWines'
deals_edu_wines = df.groupby(['NumDealsPurchases', 'Education'])['MntWines'].mean().reset_index()
deals_edu_wines
```

### üü• Create a grouped bar chart to visualize the average wine spending

```python
 
plt.figure(figsize=(15, 8))
sns.barplot(
    x='NumDealsPurchases',
    y='MntWines',
    hue='Education',
    data=deals_edu_wines,
    palette='viridis'
)
plt.title('Average Wine Spending by Number of Deal Purchases and Education Level', fontsize=16)
plt.xlabel('Number of Deals Purchased', fontsize=12)
plt.ylabel('Average Spending on Wines', fontsize=12)
plt.legend(title='Education')
plt.tight_layout()
plt.savefig('mnt_wines_by_deals_and_education.png')
plt.show()
```

### üü• Here are the key observations:

### üî¥ Impact of Deals on Wine Spending: 
The chart shows a general trend where customers who make fewer deal purchases tend to have higher average spending on wines. 
For most education levels, the average spending on wines is highest for customers who make 0 to 3 deal purchases and declines as the number of deals increases.

### üî¥ Education Level Differences:
There are variations across different education levels. For customers with the 'Master's' and 'PhD' education levels, the average spending on wines is consistently higher compared to other education levels for the same number of deal purchases. 
This suggests that highly educated customers may be more inclined to spend more on wine, regardless of deals.

### üî¥ Anomalies: 
There are some exceptions to the general trend, particularly for higher numbers of deal purchases. 
For example, customers who made a high number of deal purchases (e.g., 15) show a significant increase in average wine spending, but these data points are likely based on a very small number of customers and may not represent a broader trend.

### üü• ANALYSING NUMBER DEAL PURCHASES MADE BY YEAR_BIRTH AND EDUCATION

```python
# Group by 'Education' and 'Year_Birth' and calculate the mean of 'NumDealsPurchases'
deals_by_edu_year = df.groupby(['Education', 'Year_Birth'])['NumDealsPurchases'].mean().reset_index()

# Sort the results by 'NumDealsPurchases' in descending order to highlight the top groups
deals_by_edu_year = deals_by_edu_year.sort_values(by='NumDealsPurchases', ascending=False)

print("Top 10 Average Deal Purchases by Education and Year of Birth:")
print(deals_by_edu_year.head(10).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Create a grouped bar chart to visualize the average deal purchases by education and year of birth
plt.figure(figsize=(20, 10))
sns.barplot(
    x='Year_Birth',
    y='NumDealsPurchases',
    hue='Education',
    data=deals_by_edu_year,
    palette='viridis'
)
plt.title('Average Number of Deal Purchases by Year of Birth and Education', fontsize=16)
plt.xlabel('Year of Birth', fontsize=12)
plt.ylabel('Average Number of Deals Purchased', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Education')
plt.tight_layout()
plt.savefig('avg_deals_by_year_and_education.png')
plt.show()
```

### üî¥ Summary of Deal Purchases by Year of Birth and Education

* The analysis of NumDealsPurchases across different Year_Birth and Education levels reveals the following insights:

* The average number of deal purchases varies significantly across both Year_Birth and Education groups.

* The top average deal purchases are seen in the '2n Cycle' education group for the birth year 1967, followed by the 'Basic' education group for the birth year 1958.

Across most birth years, the number of deal purchases tends to be between 1 and 3, but there are notable peaks in specific groups, as shown in the table and chart.
